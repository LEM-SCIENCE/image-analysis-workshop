# Filtering

### Import libraries
For this example we use the **Image** and **PyPlot** modules from the Pillow (PIL) and Matplotlib libraries, respectively.

We also define a `display_at()` function to display images at different scales.

```python
from PIL import Image
import matplotlib.pyplot as plt

from lemscience.imageanalysis import show_image
```

### Read image file

```python
# Read image file and convert to grayscale
img = Image.open('../data/mandrill.png').convert('L')

# Print image mode and dimensions
print(img.mode, img.width, img.height)
```

### Display image

Use the `show_image()` function from the **imageanalysis** module to display the image. The function accepts the following optional parameters:
- `size=(width, height)`: change the image size (pixels)
- `scale=1.0`: scale the image up or down by a factor
- `grid=True|False`: overlay a grid
- `values=True|False`: overlay pixel values

```python
# Display the image at 2x enlargment
show_image(img, scale=2.0)

# Display the grid
show_image(img, scale=2.0, grid=True)

# Focus on a small part of the image
img_crop = img.crop((168, 50, 184, 66))

show_image(img_crop, size=(1000, 1000), values=True)
```

Let us no focus on a small area of the image to understand what happens during filtering is doing:

```python
img_crop = img.crop((168, 50, 184, 66))

show_image(img_crop, size=(1000, 1000), values=True)
```
